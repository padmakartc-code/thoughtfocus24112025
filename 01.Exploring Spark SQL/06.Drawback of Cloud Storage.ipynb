#Pre Requisites


## Load User_001.csdv into DataFrame 

```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_001.csv",
    header=True,
    inferSchema=True,
)
df.display()
```

#Writing into parquet format

```python
df.write.format("parquet").save(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/output_parquet",
    mode="overwrite",
)
spark.read.parquet(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/output_parquet"
).display()
```

```python
df_parquet = spark.read.parquet(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/output_parquet"
)
df_parquet.display()
```

#Ceating a View 

```python
df_parquet.createOrReplaceTempView("user_vw")
```

```python
%sql
Select * from user_vw
```

#Disadvantages - Updates are not allowed

```python
%sql
update user_vw
set country = 'Bharat'
where country = 'India'
```
