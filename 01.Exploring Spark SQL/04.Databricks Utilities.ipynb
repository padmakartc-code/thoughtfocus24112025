```python
dbutils.widgets.combobox("top_n-records",
                         defaultValue="1",
                         choices=["1", "2", "3", "4", "5"],
                         label="Top N Records")
```

```python
dbutils.widgets.get("top_n-records")
```

```python
dbutils.widgets.dropdown(name="environment",
                         defaultValue="dev",
                         choices=["dev", "test", "prod"],
                         label="Environment Type")
```

```python
dbutils.widgets.multiselect(name="user_input",
                            defaultValue="2G",
                            choices=["2G", "3G", "4G","5G"],
                            label="Technology")
```

```python
print(dbutils.widgets.get("top_n-records"),
dbutils.widgets.get("environment"))
```
