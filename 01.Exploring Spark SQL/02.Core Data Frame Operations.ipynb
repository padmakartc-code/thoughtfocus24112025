#PREREQUISITES


##Load employee.csv into DataFrame

```python
df = spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv",
                    sep="|",
                    inferSchema=True,
                    header=True,
                    quote="'")
display(df)  
```

#select operation

```python
df.select("id","name").display()
```

```python
from pyspark.sql.functions import col #recomended way to use col("column_name")
df.select(col("id"),col("name")).display()
```

```python
from pyspark.sql.functions import col
df.select(col("id").alias("Employee ID"),col("name")).display()
```

```python
df.filter(col("desig")=="Developer").display()
```

```python
df.filter((col("desig")=="Developer") | (col("desig")=="Team Lead")).display()
```

```python
df.filter((col("desig")=="Developer") | (col("gen")=="M")).display()
```

```python
df.filter(col("exp").between(1,6)).display()
```

```python
df.filter((col("exp").between(1,6)) | (col("gen")=="M")).select(col("name").alias("Employee Name")).display()
```

# Grouping Aggregations and Sorting

```python
df.groupBy(col("gen")).count().display()
```

```python
df.groupBy(col("gen")).count().sort(col("count")).display()
```

```python
df.groupBy(col("gen")).count().sort(col("count"), ascending = False).display()
```

```python
df.groupBy(col("gen")).count().sort(col("count").desc()).display()
```

```python
from pyspark.sql.functions import avg, sum, min, max
df.groupBy(col("desig")).agg(min("exp").alias("min_exp"),
                             max("exp"),
                             avg("exp")).display()
```

#Derive New Columns

```python
from pyspark.sql.functions import lit
df.withColumn("is_employeed", lit("Yes")).display()
```

```python
from pyspark.sql.functions import when
df.withColumn("category", when(col("exp")>=10, "Senior")
                               .when(col("exp")>=5, "Mid Level")
                               .when(col("exp")>=1, "Junior")
                               .otherwise("Invalid Experience")
              
              ).display()
```

#View

```python
df.createOrReplaceTempView("employee_vw")
```

```python
%sql

select * from employee_vw;
```

```python
%sql

select *, CASE WHEN exp>=10 THEN 'Senior' WHEN exp>=5 THEN 'Mid' ELSE 'Junior' END as Category from employee_vw;'
```
